```python
quik_method = test_df.copy()

quik_method = quik_method.reset_index()

se_obj = quik_method.speaker
to_work_on =  (list(se_obj))


count = 1
magic_list = []
for index,element in enumerate(to_work_on):
    if index != (len(to_work_on)-1):
        if to_work_on[index] == to_work_on[index+1] :
            magic_list.append(count)
        
        else:
            magic_list.append(count)
            count += 1
    elif index == (len(to_work_on)-1):
        if to_work_on[index-1] == to_work_on[index]:
            magic_list.append(count)
        else:
#             
            magic_list.append(count)
            
quik_method['magic_list'] = magic_list
```

```python
quik_method['new_list'] = zip(quik_method.speaker,quik_method.magic_list)
```
